```python
import time
import sys
l = range(1000)
print(sys.getsizeof(l)*len(l))
```

```python
arr = np.arange(1000)
print(arr.size*arr.itemsize)
```

```python
l1  = range(1000000)
l2  = range(1000000)
start = time.time()
result = [(x+y) for x,y in zip(l1,l2)]
print("List takes time: ",(time.time()-start)*1000)
```

```python
arr1  = np.arange(1000000)
arr2  = np.arange(1000000)
startArray = time.time()
resultArray = arr1 + arr2
print("Numpy array takes time: ",(time.time()-startArray)*1000)
```
