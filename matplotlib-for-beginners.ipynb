```python
##Matplotlib is a 2d plotting library i.e.used for data visualization 
#Matplotlib inline is a backend command in Jupyter Notebooks that enables the rendering of Matplotlib plots directly below code cells. This means that when you execute code that generates Matplotlib plots, the plots are displayed inline within the notebook, rather than in a separate window or output cell.
```

```python
import matplotlib.pyplot as plt
%matplotlib inline 
x = [1,2,3,4,5,6,7]
y = [50,51,52,48,47,49,46]
plt.plot(x,y)
```

```python
# To hide this --> [<matplotlib.lines.Line2D at 0x79b7b1ea6ef0>]
plt.plot(x,y,marker = "o",color="g",linewidth=5)
plt.show()
```

```python
#Set the linewidth and line style
plt.plot(x,y,marker = "D",color="g",linewidth=5,linestyle = "dashdot")
plt.show()
```

```python
plt.plot(x,y,marker = "*")
plt.show()
```

```python
#Set the marker size
plt.plot(x,y,marker = "d",markersize=8)
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')
plt.show()
```

```python
#Make the line transparent using alpha
plt.plot(x,y,marker = "d",markersize=8,alpha = 0.2)
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')
plt.show()
```

```python
#Plotting three lines 
days=[1,2,3,4,5,6,7]
max_t=[50,48,47,49,46,51,52]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]
plt.plot(days,max_t,color="r")
plt.plot(days,min_t,color="c") # color = "c" --> cyan and m --> magenta
plt.plot(days,avg_t,color="m")
plt.show()
```

```python
#Set the legend
plt.plot(days,max_t,color="r",label = 'Max')
plt.plot(days,min_t,color="c",label = 'Min') # color = "c" --> cyan and m --> magenta
plt.plot(days,avg_t,color="m",label = 'Avg')
plt.legend(loc="best",shadow=True,fontsize='small') #loc="best" --> It will automatically shift its position
plt.show()
```

# Histogram

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
plt.hist(blood_sugar,color='m')
plt.show()
```

```python
#Set histogram relative width and histtype
plt.hist(blood_sugar,color='m',rwidth = 0.6,histtype="step")
plt.show()
```

**Bins, or buckets, are the ranges of numeric values that each bar in a histogram represents. The height of the bar indicates how many data points fall within that bin.**

```python
blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]
plt.hist([blood_sugar_men,blood_sugar_women],color=['blue','orange'],label = ['Men','Women'],rwidth = 0.9)
plt.xlabel('sugar range')
plt.ylabel('Total no of patient')
plt.title('Blood Sugar Analysis')
plt.legend(loc='best')
plt.show()
```

# **Make historgram horizontal**

```python
plt.hist([blood_sugar_men,blood_sugar_women],color=['blue','orange'],label = ['Men','Women'],rwidth = 0.9,orientation = 'horizontal')
plt.xlabel('sugar range')
plt.ylabel('Total no of patient')
plt.title('Blood Sugar Analysis')
plt.legend(loc='best')
plt.show()
```

# Pie Chart


```python
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
mycolors = ["c", "hotpink", "r", "m"]
plt.pie(exp_vals,labels = exp_labels,colors = mycolors)
plt.legend()
plt.show()
```

```python
# Make pie stand out
myexplode = [0,0,0.2,0,0]
plt.pie(exp_vals,labels = exp_labels,colors = mycolors,explode=myexplode)
plt.legend(loc = 'upper right')
plt.show()
```

```python
#Calculate the percentage of pie
plt.pie(exp_vals,labels = exp_labels,colors = mycolors,explode=myexplode,autopct = "%0.2f%%")
plt.legend(loc = 'upper right')
plt.show()
```

```python
# Set the shadow of pie
plt.pie(exp_vals,labels = exp_labels,colors = mycolors,explode=myexplode,autopct = "%0.2f%%",shadow = True)
plt.legend()
plt.show()
```

```python
#Increase the radius of pie
plt.pie(exp_vals,labels = exp_labels,colors = mycolors,explode=myexplode,autopct = "%0.2f%%",radius = 1.5,shadow=True)
plt.legend()
plt.show()
```

```python
# Set the startangle of pie
plt.pie(exp_vals,labels = exp_labels,colors = mycolors,explode=myexplode,autopct = "%0.2f%%",radius = 1.5,shadow=True,startangle = 45)
plt.legend()
plt.show()
```

```python
# Save the piechart image
plt.pie(exp_vals,labels = exp_labels,colors = mycolors,explode=myexplode,autopct = "%0.2f%%",radius = 1.5,shadow=True,startangle = 45)
plt.legend()
plt.savefig('piechart.png',bbox_inches = 'tight',pad_inches = 2,transparent = True)
plt.show()

#It will save the piechart image in PNG format.
#bbox_inches --> it will fit the pie chart image.
#pad_inches --> It will create more space from left to right & top to bottom 
```
